## Part 1:

## Video Game Sales

Memuat set data :

```python
import pandas as pd
import numpy as np
vidgame = pd.read_csv('vgsales.csv')
vidgame
```

### Temukan video game yang menghasilkan pendapatan terbanyak secara global

```python
vidgame.sort_values('Global_Sales', ascending = False).head(1)['Name']
```

### Temukan platform dengan penjualan global paling rata-rata.

```python
platforms = vidgame.groupby('Platform')
platforms.mean().sort_values('Global_Sales',ascending = False).head(1)
```

### Temukan penerbit dengan penjualan global terbanyak

```python
publishers = vidgame.groupby('Publisher')
publishers.sum().sort_values('Global_Sales', ascending = False).head(1)
```

### Menemukan tahun mana yang memiliki penjualan global paling banyak

```python
a = vidgame.groupby('Year')
a.sum().sort_values('Global_Sales', ascending = False).head(1)
```

### Menemukan game terlaris untuk setiap platform

```python
game_terlaris = vidgame.groupby(['Platform','Name','Global_Sales'])
game_terlaris.max().sort_values('Global_Sales', ascending = False).groupby('Platform').head()
```

### Menemukan penjual teratas untuk setiap genre

```python
vidgame.groupby(['Genre','Name','Global_Sales']).max().sort_values('Global_Sales', ascending = False).groupby('Genre').head(1)
```

```python
penjualan_teratas = vidgame.groupby(['Genre','Name','Global_Sales'])
penjualan_teratas.max().sort_values('Global_Sales', ascending = False).groupby('Genre').head(1)
```

### Membuat fungsi yang mengelompokkan tanggal rilis video game berdasarkan dekade, dan menunjukkan bagaimana penjualan global mereka telah berevolusi

```python
def decades(row):
    year = row['Year']
    
    if 1979 < year < 1990:
        return "1980s"
    elif 1989 < year < 2000:
        return "1990s"
    elif 1999< year < 2010:
        return "2010s"
    else:
        return "others"

vidgame['Decade'] = vidgame.apply(decades, axis = 'columns')
vidgame.head(20)
```

```python
#import library
import pandas as pd
import numpy as np
```

Memuat dan melihat kumpulan data:

```python
#memuat dataset
movie = pd.read_csv('movie_metadata.csv')
```

```python
#dataset 5 teratas
movie.head()
```

### Menemukan perbedaan antara anggaran dan total box office per sutradara setiap tahun dengan menambahkan kolom baru bernama` diff_gross`.

```python
# drop missing values
df = movie.copy()
df = df.dropna(how = 'any')
df.head()
```

```python
#kolom baru diff gross
df['diff_gross'] = df['gross'] - df['budget']
df.head()
```

```python
# lakukan group by diff gross kemudian urutkan secara descending
directors = df.groupby('director_name')
directors.sum().sort_values('diff_gross', ascending = False).head()
```

### Menemukan direktur mana yang memperoleh jumlah ulasan kritikus tertinggi

```python
kritik_tertinggi = movie.groupby('director_name')
kritik_tertinggi.sum().sort_values('num_critic_for_reviews', ascending = False).head(1)
```

### IMDB dikenal karena ketidakseimbangan antara evaluasi kritik profesional dan pengguna platform; Nilai tinggi berarti kritikus profesional lebih menghargai film dan sebaliknya. Saya akan menemukan rasio ini dan menambahkannya di kolom baru bernama `crit_ratio`.

```python
df['critic_ratio'] = df['num_critic_for_reviews'] / df['num_user_for_reviews']
df.head()
```

### Temukan 20 sutradara terbaik menurut rasio kritikus

```python
new = df.copy()
new.set_index('director_name', inplace = True)
```

```python
director = df.groupby('director_name')
director.mean().sort_values('critic_ratio', ascending=False).head(20)
```

```python
new['genre_list'] = new['genres'].str.split('|', expand = False)
new.head()
```
